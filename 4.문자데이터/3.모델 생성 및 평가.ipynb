### 모델 생성 및 평가

```python
import pandas as pd
df = pd.read_csv("data/loan_train_cleaning.csv")
df.head().T #컬럼/인덱스 좌우 위치 바꿔줌 (컬럼 많을 경우 확인 용이)
```

### 머신러닝 변수
* label : 종속 변수 또는 정답
* feature : 독립 변수 또는 속성

```python
df.columns
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']
label = 'Loan_Status'
X, y = df[features] , df[label]
```

```python
X.T
```

```python
df.shape
```

- model_selection : data 쪼개기(8:2 / 7:3)
    - test_size : test 비중
    - X_train : 과정 / y_train : 정답
    - X_test : test / y_test : 정답

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train , y_test = train_test_split(X, y, test_size=0.2)
```

- 8 : 2 로 데이터 나눠지는 것을 볼 수 있음

```python
X_train.shape, X_test.shape
```

```python
y_train.shape , y_test.shape
```

- 알고리즘에 학습 시키기

### 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려줌
- DecisionTreeClassifier : 분류 / DecisionTreeRegressor : 회귀

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
pred = dtc.predict(X_test)
pred
```

- 예측 값과 정답 비교
    - accuracy_score : 정확도 나타냄

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
